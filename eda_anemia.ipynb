```python
renombrar = {
    'Fecha_Atencion': 'fecha_atencion',
    'RENIPRESS': 'renipress',
    'IPRESS': 'ipress',
    'Id_Genero': 'genero',
    'Edad_meses': 'edad',
    'Diagnostico': 'diagnostico'
}
```

```python
df = df.rename(columns=renombrar)
df.head()
```

```python
df['fecha_atencion'] = pd.to_datetime(df['fecha_atencion'], errors='coerce', dayfirst=True)
```

```python
df[['renipress', 'ipress', 'genero', 'diagnostico']] = df[['renipress', 'ipress', 'genero', 'diagnostico']].astype('string')
```
